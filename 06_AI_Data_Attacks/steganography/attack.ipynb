```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
import numpy as np
import os

SEED = 1337
np.random.seed(SEED)
torch.manual_seed(SEED)


# Define a simple Neural Network
class SimpleNet(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SimpleNet, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, output_size)
        # Add a larger layer potentially suitable for steganography later
        self.large_layer = nn.Linear(hidden_size, hidden_size * 5)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        # Note: large_layer is defined but not used in forward pass for simplicity
        # In a real model, all layers would typically be used.
        return x


# Model parameters
input_dim = 10
hidden_dim = 64  # Increased hidden size for larger layers
output_dim = 1  # e.g., for regression or binary classification sigmoid output
target_model = SimpleNet(input_dim, hidden_dim, output_dim)

print("SimpleNet model structure:")
print(target_model)
print("\nModel parameters (state_dict keys):")
for name, param in target_model.state_dict().items():
    print(f"  {name}: shape={param.shape}, numel={param.numel()}")
```

```python
# Generate dummy data
num_samples = 100
X_train = torch.randn(num_samples, input_dim)
# Simple linear relationship + noise for dummy target
true_weights = torch.randn(input_dim, output_dim)
y_train = X_train @ true_weights + torch.randn(num_samples, output_dim) * 0.5

# Create DataLoader
dataset = TensorDataset(X_train, y_train)
dataloader = DataLoader(dataset, batch_size=16)

# Loss and optimizer
criterion = nn.MSELoss()  # Mean Squared Error loss
optimizer = optim.Adam(target_model.parameters(), lr=0.01)

# Simple training loop
num_epochs = 5  # Just a few epochs
print(f"\n'Training' the model for {num_epochs} epochs...")
target_model.train()  # Set model to training mode
for epoch in range(num_epochs):
    epoch_loss = 0.0
    for inputs, targets in dataloader:
        optimizer.zero_grad()  # Clear previous gradients
        outputs = target_model(inputs)  # Forward pass
        loss = criterion(outputs, targets)  # Calculate loss
        loss.backward()  # Backpropagation
        optimizer.step()  # Update weights
        epoch_loss += loss.item()
    print(
        f"  Epoch [{epoch + 1}/{num_epochs}], Loss: {epoch_loss / len(dataloader):.4f}"
    )

print("Training complete.")
```

```python
import struct  # for float packing/unpacking


# ──────────────────────────────────────────────────────────────
# TODO [STUB 1/5]: implement encode_lsb
#   Encodes byte data into the LSBs of a float32 tensor.
#   Hints:
#     - Prepend a 4-byte big-endian length header before data_bytes
#     - Bit-stream the payload, pack `num_lsb` bits per element
#     - Reinterpret float32 as uint32, clear the bottom num_lsb bits,
#       then OR in the payload bits
#     - Return a NEW tensor (don't mutate tensor_orig)
# ──────────────────────────────────────────────────────────────
def encode_lsb(
    tensor_orig: torch.Tensor, data_bytes: bytes, num_lsb: int
) -> torch.Tensor:
    """Encodes byte data into the LSBs of a float32 tensor (prepends length)."""
    if tensor_orig.dtype != torch.float32:
        raise TypeError("Tensor must be float32")
    if not 1 <= num_lsb <= 8:
        raise ValueError("num_lsb must be 1-8")
    
    tensor = tensor_orig.clone().detach()

    num_els = tensor.numel()
    tensor_flat = tensor.flatten()

    data_len = len(data_bytes)
    data_to_embed = struct.pack(">I", data_len) + data_bytes
    print(f"Type: {type(data_to_embed)}")
    total_bits_needed = len(data_to_embed) * 8
    capacity_bits = num_els * num_lsb

    if total_bits_needed > capacity_bits:
        raise ValueError(f"Tensor too small: needs {total_bits_needed}, has {capacity_bits} bits.")

    data_iter = iter(data_to_embed)
    current_byte = next(data_iter, None)

    bit_index_in_byte = 7 
    element_index = 0
    bits_embedded = 0

    while bits_embedded < total_bits_needed and element_index < num_els:
        if current_byte is None:
            break

        orig_float = tensor_flat[element_index].item()
        packed_float = struct.pack(">f", orig_float)
        int_repr = struct.unpack(">I", packed_float)[0]

        mask = (1 << num_lsb) - 1
        bits_to_add = 0

        for i in range(num_lsb):
            data_bit = (current_byte >> bit_index_in_byte) & 1
            bits_to_add |= data_bit << (num_lsb-1-i)

            bit_index_in_byte -= 1
            if bit_index_in_byte < 0:
                current_byte = next(data_iter, None)
                bit_index_in_byte = 7

            bits_embedded += 1
            if bits_embedded >= total_bits_needed:
                break
        
        cleared_bit = int_repr & (~mask)
        new_int_repr = cleared_bit | bits_to_add

        new_packed_float = struct.pack(">I", new_int_repr)
        new_float = struct.unpack(">f", new_packed_float)[0]

        tensor_flat[element_index] = new_float
        element_index += 1
    
    return tensor

    
    


# ──────────────────────────────────────────────────────────────
# TODO [STUB 2/5]: implement decode_lsb
#   Decodes byte data hidden in tensor LSBs.
#   Hints:
#     - First 32 bits across ceil(32/num_lsb) elements → length L (big-endian)
#     - Then read next L*8 bits from following elements
#     - Reassemble into bytes and return
# ──────────────────────────────────────────────────────────────
def decode_lsb(tensor_modified: torch.Tensor, num_lsb: int) -> bytes:
    """Decodes byte data hidden in tensor LSBs (reads length first)."""
    if tensor_modified.dtype != torch.float32:
        raise TypeError("Tensor must be float32")
    if not 1 <= num_lsb <= 8:
        raise ValueError("num_lsb must 1-8")
    
    tensor_flat = tensor_modified.flatten()
    n_elems = tensor_flat.numel()
    curr_float_ix = 0


    def get_bits(count):
        nonlocal curr_float_ix
        bits = []

        while len(bits) < count and curr_float_ix < n_elems:
            curr_float = tensor_flat[curr_float_ix].item()

            pack_float = struct.pack(">f", curr_float)
            n_repr = struct.unpack(">I", pack_float)[0]

            mask = (1 << num_lsb) -1
            extracted_bits = n_repr & mask

            for i in range(num_lsb):
                bit = (extracted_bits >> (num_lsb - 1 - i)) & 1
                bits.append(bit)

                if len(bits) >= count:
                    break
            curr_float_ix += 1
        if len(bits) < count:
            raise ValueError("Tensor ended prematurely.")
        
        return bits

    try:
        length_bits = get_bits(32)
    except ValueError as e:
        raise ValueError("Failed length decode: " + str(e))
    
    payload_len = 0
    for b in length_bits:
        payload_len = (payload_len << 1) | b
    
    if payload_len == 0:
        return b""
    
    try:
        payload_bits = get_bits(payload_len*8)
    except ValueError as e:
        raise ValueError("Failed payload decode: " + str(e))
    
    payload_bytes = bytearray()
    curr_byte = 0
    bit_count = 0
    for bit in payload_bits:
        curr_byte = (curr_byte << 1) | bit
        bit_count += 1

        if bit_count == 8:
            payload_bytes.append(curr_byte)
            curr_byte = 0
            bit_count = 0
    
    return bytes(payload_bytes)
    
print("Steganography functions encode_lsb() and decode_lsb() defined.")
```
