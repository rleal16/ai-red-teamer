```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.DataFrame(
    {
        "column1": np.random.rand(50),
        "column2": np.random.rand(50)*10,
    }
)

print(data.head())

plt.scatter(data["column1"], data["column2"])
plt.xlabel("Column 1")
plt.ylabel("Column 2")
plt.title("Scatter Plot")
plt.show()
```
